```python
import pandas as pd
```

```python
pok = pd.read_csv("Pokemon.csv")
```

```python
speed = pok["Speed"].mean()
print("Speed: ",speed)
pok_hp = pok["HP"].mean()
print("HP: ",pok_hp)
```

```python
def sp(value):
    if value < speed:
        return "Speed is Low. "
    elif value > speed:
        return "Speed is High. "
    else:
        return "Speed is maintained. "

def pk_hp(value):
    if value < pok_hp:
        return "HP is Low. "
    elif value > pok_hp:
        return "HP is High. "
    else:
        return "HP is maintained. "    
    
```

```python
pok["High_Low_Speed"] = pok["Speed"].apply(sp)
pok["High_Low_HP"] = pok["HP"].apply(pk_hp)
pok
```
